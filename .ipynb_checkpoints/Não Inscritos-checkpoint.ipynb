```python
import pandas as pd
import numpy as np
from django_pandas.io import read_frame

a = Inscricao.objects.filter(absoluto=True)
df = read_frame(a, fieldnames=['atleta', 'atleta__faixa', 'categoria_idade', 'atleta__academia'])
pv = df.pivot_table(index=['atleta__academia', 'atleta'], aggfunc='count', fill_value=0, margins=True)
pv
```

```python
res = Resultado.objects.all()
dr = read_frame(res, fieldnames=['atleta', 
                                 'absoluto', 
                                 'pontos', 
                                 'atleta__atleta__academia', 
                                 'atleta__atleta__academia__equipe'])
pr = dr.pivot_table(index=['atleta__atleta__academia__equipe'], 
                    aggfunc='sum', 
                    fill_value=0, 
                    margins=True)
pa = dr.pivot_table(index=['atleta__atleta__academia'], 
                    aggfunc='sum', 
                    fill_value=0, 
                    margins=True)
```
